```python
import matplotlib.pyplot as plt
import tensorflow as tf
from skimage import io
from skimage.transform import resize
import numpy as np
import cv2
import os

# Load the saved model
model = tf.keras.models.load_model('model_for_nuclei.keras', custom_objects={'Lambda': tf.keras.layers.Lambda}, compile=False, safe_mode=False)


# Load your test data (replace this with your actual test data)
path="C:\\Users\\Mayank\\Downloads\\VineNet\\VineNet\\data\\test\\images"
path2="C:\\Users\\Mayank\\Downloads\\VineNet\\VineNet\\data\\test\\masks"

for images in os.listdir(path):
    img = io.imread(os.path.join(path,images))


    test_data = np.expand_dims(img, axis=0)


    pred = model.predict(test_data)
    pred = pred[0]



# Convert predicted mask to binary
    threshold = 0.17 # Define the threshold value
    binary_mask = np.where(pred > threshold, 1, 0)



# Convert the binary mask to uint8 for visualization
    binary_mask_uint8 = (binary_mask * 255).astype(np.uint8)
    plt.imshow(binary_mask_uint8)
    cv2.imwrite(os.path.join(path2,images),binary_mask_uint8)
    




```
